```python
import tkinter as tk
from IPython.display import display, HTML

# Function to change font style
def change_font():
    # In a Colab environment, we can't directly modify the Tkinter label
    # Instead, we'll display the text with the desired font style
    display(HTML("<p style='font-family: Arial; font-size: 18px; font-weight: bold;'>Hello, World!</p>"))

# Since we cannot run a full Tkinter app in Colab, we will simulate the output

# Initial display of the text
display(HTML("<p style='font-family: Helvetica; font-size: 14px;'>Hello, World!</p>"))

# Call the function to show how the text would be displayed with the changed font
print("Simulating button click to change font:")
change_font()
```
