```python
def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def multiply(a, b):
    return a * b
def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero")
    return a / b

a = 10
b = 5
print("Addition:", add(a, b))
print("Subtraction:", subtract(a, b))
print("Multiplication:", multiply(a, b))
print("Division:", divide(a, b))
```

```python
def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def multiply(a, b):
    return a * b
def divide(a, b):
    if b == 0:
        return "Error! Division by zero."
    return a / b
import math
def circle_area(radius):
    return math.pi * radius * radius
def rectangle_area(length, width):
    return length * width
def triangle_area(base, height):
    return 0.5 * base * height
print("Circle Area (radius=7):", circle_area(7))
print("Rectangle Area (5x10):", rectangle_area(5, 10))
print("Triangle Area (base=6, height=8):", triangle_area(6, 8))
```
